```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diff_filename = "percentage_difference.csv"
diff_df = pd.read_csv(diff_filename) 
```

```python
array_percentage = diff_df['diff_percentage']
```

```python
sites = [i+1 for i in range(49)]
sites = [str(x) for x in sites]

height = array_percentage

fig = plt.figure(figsize = (38, 15))
plt.rcParams.update({'font.size': 28})

# creating the bar plot
plt.bar(sites, height, color ='orange',
        width = 0.8
    )

 
plt.xlabel('Site', fontsize=35)
plt.ylabel('Proportional difference (%)',fontsize=35)
#plt.title("Students enrolled in different courses")
plt.show()
fig.savefig('ML_area_difference.png',bbox_inches='tight')
```
